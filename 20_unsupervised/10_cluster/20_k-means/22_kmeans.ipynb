```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 系统库
import os, sys

# 自带数据
datalib_path = os.path.join(os.path.abspath('.'), '../../../../')
sys.path.append(datalib_path)
import dataset
```

```python
dataSet = []
fr = open(os.path.join(dataset.cluster_path, 'testSet.txt'))
for line in fr.readlines():
    curLine = line.strip().split('\t')
    dataSet.append([float(curLine[0]), float(curLine[1])])
```

```python
km = KMeans(n_clusters=4)
km.fit(dataSet)
km_pred = km.predict(dataSet)
centers = km.cluster_centers_
```

```python
plt.scatter(np.array(dataSet)[:, 1], np.array(dataSet)[:, 0], c=km_pred)
plt.scatter(centers[:, 1], centers[:, 0], c="r")
plt.show()
```
